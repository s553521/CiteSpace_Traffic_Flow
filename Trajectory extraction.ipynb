```python
import pandas as pd
```

```python
#Data preview
test = pd.read_csv('..\\test_data.csv')
test
```

```python
#Trajectory extraction
test['od'] = 0
i = 0
x = 0
while i < len(test.index):
    try:
        if test.loc[0,'cus'] == 1:
            x += 1
        elif test.loc[i].values[5:6] == 0 and test.loc[i-1].values[5:6] == 1 and test.loc[i-2].values[5:6] == 1:
            test.loc[i-x, 'od'] = 1
            test.loc[i-1, 'od'] = 2
            x = 0
        elif test.loc[i].values[5:6] == 0:
            x = 0
        i += 1
    except Exception as e:
        i += 1
```

```python
import pandas as pd
import time
```

```python
#Data preview
test = pd.read_csv("..\\Data\\test_data_name_transformed.csv")
test
```

```python
#Trajectory get
test['de'] = str('')
i = 0
x = 0
while i < len(test.index):
    if test.loc[i, 'od'] == 1:
        x = i
        test.loc[x, 'de'] += str(test.loc[i, 'road_name']) + ';'
        i += 1
    if test.loc[i, 'od'] == 2:
        test.loc[x, 'de'] += str(test.loc[i, 'road_name']) + '+'
        i += 1
        continue
    test.loc[x, 'de'] += str(test.loc[i, 'road_name']) + '+'
    i += 1
test = test.loc[test['road_name'] != '']
```

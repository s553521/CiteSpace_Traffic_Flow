```python
import pandas as pd
```

```python
#The data calculated from CiteSpace
test = pd.read_csv('..\\Data\\test_20140808_1998.csv')
test.drop('Unnamed: 0', axis = 1, inplace = True)
test
```

```python
#Road classes data preview
road = pd.read_csv('..\\Data\\test_road_class.txt', sep = ',')
road
```

```python
#Road classes match roads
i = 0
j = 0
test['funclass'] = 0
while i < len(test.index):
    j = 0
    while j < len(road.index):
        if test.loc[i, 'funclass'] == 0:
            if test.loc[i, 'name'] == road.loc[j, 'PathName']:
                test.loc[i, 'funclass'] = road.loc[j, 'FuncClass']
                j = len(road.index)
        if test.loc[i, 'funclass'] != 0:
            j = len(road.index)
        j += 1
    i += 1
```

```python
test.to_csv('..\\Data\\test_data_for_mining', encoding = 'utf-8-sig', index = False)
```
